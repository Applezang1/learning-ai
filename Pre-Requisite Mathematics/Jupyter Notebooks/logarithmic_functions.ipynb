# Logarithmic Functions 
This notebook aims to help gain a better understanding of logarithmic functions by defining a logarithmic function, plotting the function, and finding the output values for the function with different input values.

### Imports
Import the libraries necessary to define the logarithmic function and graph it.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define Input Values 
Define a input variable (x) as an array of values from -5 to 5 with increments of 0.01 between each value.

```python
x = np.arange(0.01,5.0, 0.01)
```

### Define the Logarithmic Function 
Define the logarithmic function as an input of the variable x and store the corresponding output in the output variable (y)

```python
def logarithmic_function(x): 
   y = np.log(x)
   return y 
```

### Plot the Logarithmic Function
Plot the x values and the corresponding y values on a Cartesian coordinate system.

```python
fig, ax = plt.subplots()
ax.plot(x,logarithmic_function(x),'r-')
ax.set_ylim([-5,5]);ax.set_xlim([0,5])
ax.set_xlabel('x'); ax.set_ylabel('$\\log[x]$')
plt.show()
```

### Compute the Logarithmic Function 
Compute the logarithmic function using different x values (0.1, 1, e, e^3) and print the output of the logarithmic function for each value of x.

```python
print(f'The output of the logarithmic function when x = 0.1 is {logarithmic_function(0.1)}')
print(f'The output of the logarithmic function when x = 1 is {logarithmic_function(1)}')
print(f'The output of the logarithmic function when x = e is {logarithmic_function(np.e)}')
print(f'The output of the logarithmic function when x = e^3 is {logarithmic_function(np.e**3)}')
```
