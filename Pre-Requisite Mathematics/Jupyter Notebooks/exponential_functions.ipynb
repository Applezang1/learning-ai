# Exponential Functions 
This notebook aims to help gain a better understanding of exponential functions by defining an exponential function, graphing it, and finding the output of the exponential functions with different input values.

### Imports
Import the libraries needed to define and plot an exponential function

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define Input Values
Define the input variable (x) as an array of values from -5 to 5 with increments of 0.01 for each value

```python
x = np.arange(-5.0,5.0, 0.01)
```

### Define the Exponential Function
Define the exponential function as a function of the input variable (x) and store the value in the output variable (y).

```python
def exponential_function(x): 
   y = np.exp(x)
   return y 
```

### Plot the Function 
Plot the x values and the corresponding y values onto a Cartesian coordinate system.

```python
fig, ax = plt.subplots()
ax.plot(x,exponential_function(x),'r-')
ax.set_ylim([0,100]);ax.set_xlim([-5,5])
ax.set_xlabel('x'); ax.set_ylabel('exp[x]')
plt.show() 
```

### Calculate Exponential Function 
Calculate the exponential function with different input values (0, 1, infinity, negative infinity) and print the resulting output.

```python
print(f'The output of the exponential function when x = 0 is {exponential_function(0)}')
print(f'The output of the exponential function when x = 1 is {exponential_function(1)}')
print(f'The output of the exponential function as x approaches infinity is {exponential_function(np.inf)}')
print(f'The output of the exponential function as x approaches negative infinity is {exponential_function(-np.inf)}')
```
