# Shallow Network (1_1_3)
This notebook aims to get a better understanding of shallow neural networks by defining a shallow neural network with 1 input, 1 output, and 3 hidden units, plotting it, and computing the least squares loss function of the shallow neural network in comparison to the given input/output data pairs

## Definition of Shallow Neural Network 
**Shallow neural network** is a type of neural network with one hidden layer. A shallow neural network consists of the preactivation, activation, weighted activation, and output. 

**Preactivation**: The preactivation is a function in terms of a number of parameter and the input value 

**Activation**: The activation is the output of the preactivation after it is inputted through an activation function (ReLU for example) 

**Weighted Activation**: The weighted activation is the activation offsetted/multipled by a parameter 

**Output**: The output is the combination of the weighted activation values plus an offset by a parameter

### Imports 
Import the libraries needed to define the shallow neural network and plot it

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Rectified Linear Unit Function (ReLU) 
A **rectified linear unit function (ReLU)** is a function where all negative output values are converted to 0. ReLU is an activation function that is applied to the preactivation function to allow the shallow neural network to model nonlinear relationships. 

### Define the Rectified Linear Unit (ReLU) function 
Define the rectified linear unit function as a function that takes preactivation as its input and outputs the activation value.

```python
def ReLU(preactivation):
  activation = np.clip(preactivation, 0, None)
  return activation
```

### Define Input Values for ReLU function
Define the input value (z) as an array of numbers from -5 to 5 with increment 0.1 between each value

```python
z = np.arange(-5,5,0.1)
```

### Compute the ReLU function
Compute the ReLU function by using the defined input values (z) and inputting it into the ReLU function. Store the outputs into a variable called RelU_z.

```python
RelU_z = ReLU(z)
```

### Plot the ReLU function

```python
fig, ax = plt.subplots()
ax.plot(z,RelU_z,'r-')
ax.set_xlim([-5,5]);ax.set_ylim([-5,5])
ax.set_xlabel('z'); ax.set_ylabel('ReLU[z]')
ax.set_aspect('equal')
plt.show()
```

## Shallow Neural Networks

### Define the Shallow Neural Network 
Define a shallow neural network with one input (x), one output (y), three hidden units, and a set of parameters (phi_0, phi_1, phi_2, phi_3, theta_10, theta_11, theta_29, theta_21, theta_39, theta_31.

```python
def shallow_1_1_3(x, activation_fn, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31):
  # Define the pre activations
  pre_1 = theta_10 + theta_11 * x
  pre_2 = theta_20 + theta_21 * x
  pre_3 = theta_30 + theta_31 * x

  # Pass these through the ReLU function to compute the activations
  act_1 = activation_fn(pre_1)
  act_2 = activation_fn(pre_2)
  act_3 = activation_fn(pre_3)

  # Weight the activations using phi1, phi2 and phi3
  w_act_1 = phi_1 * act_1
  w_act_2 = phi_2 * act_2
  w_act_3 = phi_3 * act_3

  # Combine the weights and an offset (phi0) to compute the output
  y = phi_0 + w_act_1 + w_act_2 + w_act_3

  return y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3
```

### Define Plotting Function 
Define a function to plot the output of the shallow neural network as well as its intermediate plots (preactivation, activation, weighted act)

```python
def plot_neural(x, y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3, x_data=None, y_data=None):

  # Function to plot intermediate plots (preactivation, activation, weighted act)
  fig, ax = plt.subplots(3,3)
  fig.set_size_inches(8.5, 8.5)
  fig.tight_layout(pad=3.0)

  # Function to plot all the pre activations
  ax[0,0].plot(x,pre_1,'r-'); ax[0,0].set_ylabel('Preactivation')
  ax[0,1].plot(x,pre_2,'b-'); ax[0,1].set_ylabel('Preactivation')
  ax[0,2].plot(x,pre_3,'g-'); ax[0,2].set_ylabel('Preactivation')

  # Function to plot all the activation
  ax[1,0].plot(x,act_1,'r-'); ax[1,0].set_ylabel('Activation')
  ax[1,1].plot(x,act_2,'b-'); ax[1,1].set_ylabel('Activation')
  ax[1,2].plot(x,act_3,'g-'); ax[1,2].set_ylabel('Activation')

  # Function to plot all the weighted acts
  ax[2,0].plot(x,w_act_1,'r-'); ax[2,0].set_ylabel('Weighted Act')
  ax[2,1].plot(x,w_act_2,'b-'); ax[2,1].set_ylabel('Weighted Act')
  ax[2,2].plot(x,w_act_3,'g-'); ax[2,2].set_ylabel('Weighted Act')

  # Function to Display the plots for the pre activations, activation, weighted acts
  for plot_y in range(3):
      for plot_x in range(3):
        ax[plot_y,plot_x].set_xlim([0,1]);ax[plot_x,plot_y].set_ylim([-1,1])
        ax[plot_y,plot_x].set_aspect(0.5)
      ax[2,plot_y].set_xlabel('Input, x')
  plt.show()

  # Function to Plot and Display the resulting/final output
  fig, ax = plt.subplots()
  ax.plot(x,y)
  ax.set_xlabel('Input, x'); ax.set_ylabel('Output, y')
  ax.set_xlim([0,1]);ax.set_ylim([-1,1])
  ax.set_aspect(0.5)
  if x_data is not None:
    ax.plot(x_data, y_data, 'mo')
    for i in range(len(x_data)):
      ax.plot(x_data[i], y_data[i],)
  plt.show() 
```

### Define Parameters 

```python
theta_10 =  0.3 ; theta_11 = -1.0
theta_20 = -1.0  ; theta_21 = 2.0
theta_30 = -0.5  ; theta_31 = 0.65
phi_0 = -0.3; phi_1 = 2.0; phi_2 = -1.0; phi_3 = 7.0
```

### Define Input Values 
Define the input values (x) as an array from 0 to 1 with increments of 0.01 between each input value

```python
x = np.arange(0,1,0.01)
```

### Compute the Shallow Neural Network 
Compute the shallow neural network by using the given input values and the given parameters

```python
y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3 = \
    shallow_1_1_3(x, ReLU, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31)
```

### Plot the Shallow Neural Network

```python
plot_neural(x, y, pre_1, pre_2, pre_3, act_1, act_2, act_3, w_act_1, w_act_2, w_act_3)
```

## Least Squares Loss Function

### Definition of Least Squares Loss Function 
**Least squares loss function** is a function used to calculate the difference between the predicted value and the actual value by finding the summation of the squared differences between the predicted and actual value of a function

### Define the Least Squares Loss Function 
Define the least squares loss function as the squared difference between the predicted output values (y_predict) and the actual output values (y_train)

```python
def least_squares_loss(y_train, y_predict):
  loss = np.sum(y_predict - y_train)**2

  return loss
```

### Define Input and Output Values 
Define a set of input values (x_train) and its corresponding output values (y_train) as the actual output values, which will then be compared with the predicted output values using the least squares loss function

```python
x_train = np.array([0.09291784,0.46809093,0.93089486,0.67612654,0.73441752,0.86847339,\
                   0.49873225,0.51083168,0.18343972,0.99380898,0.27840809,0.38028817,\
                   0.12055708,0.56715537,0.92005746,0.77072270,0.85278176,0.05315950,\
                   0.87168699,0.58858043])
y_train = np.array([-0.15934537,0.18195445,0.451270150,0.13921448,0.09366691,0.30567674,\
                    0.372291170,0.40716968,-0.08131792,0.41187806,0.36943738,0.3994327,\
                    0.019062570,0.35820410,0.452564960,-0.0183121,0.02957665,-0.24354444, \
                    0.148038840,0.26824970])
```

### Compute the Shallow Neural Network 
Compute the shallow neural network to find the predicted y values (y_predict) using the given input values (x_train)

```python
y_predict, *_ = shallow_1_1_3(x_train, ReLU, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31)
```

### Compute the Least Squares Loss Function 
Compute the least squares loss function between the output values computed by the shallow neural network (y_predict) and the actual output values (y_train) to determine how well the shallow neural network fits the set of input and output values.

```python
loss = least_squares_loss(y_train,y_predict)
print('Loss = %3.3f'%(loss))
```
