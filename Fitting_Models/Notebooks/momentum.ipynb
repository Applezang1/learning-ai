### Define the Gabor Model
Define the Gabor model with two parameters (horizontal offset, frequency)

```python
def model(phi,x):
  sin_component = np.sin(phi[0] + 0.06 * phi[1] * x)
  gauss_component = np.exp(-(phi[0] + 0.06 * phi[1] * x) * (phi[0] + 0.06 * phi[1] * x) / 32)
  y_pred= sin_component * gauss_component
  return y_pred
```

### Define Gabor Model Plotting Function

```python
def draw_model(data,model,phi,title=None):
  x_model = np.arange(-15,15,0.1)
  y_model = model(phi,x_model)

  fig, ax = plt.subplots()
  ax.plot(data[0,:],data[1,:],'bo')
  ax.plot(x_model,y_model,'m-')
  ax.set_xlim([-15,15]);ax.set_ylim([-1,1])
  ax.set_xlabel('x'); ax.set_ylabel('y')
  if title is not None:
    ax.set_title(title)
  plt.show()
```

### Define Parameters

### Define Loss Function 
Define the loss function of the Gabor model as the least squares loss function

```python
def compute_loss(data_x, data_y, model, phi):
  pred_y = model(phi, data_x)
  loss = np.sum((pred_y-data_y)*(pred_y-data_y))
  return loss
```

### Define Gabor Derivative (phi[0]) Function
Define a function that computes the derivative of the Gabor model in respect to phi0

```python
def gabor_deriv_phi0(data_x,data_y,phi0, phi1):
    x = 0.06 * phi1 * data_x + phi0
    y = data_y
    cos_component = np.cos(x)
    sin_component = np.sin(x)
    gauss_component = np.exp(-0.5 * x *x / 16)
    deriv = cos_component * gauss_component - sin_component * gauss_component * x / 16
    deriv = 2* deriv * (sin_component * gauss_component - y)
    return np.sum(deriv)
```

### Define Gabor Derivative (phi[1]) Function
Define a function that computes the derivative of the Gabor model in respect to phi1

```python
def gabor_deriv_phi1(data_x, data_y,phi0, phi1):
    x = 0.06 * phi1 * data_x + phi0
    y = data_y
    cos_component = np.cos(x)
    sin_component = np.sin(x)
    gauss_component = np.exp(-0.5 * x *x / 16)
    deriv = 0.06 * data_x * cos_component * gauss_component - 0.06 * data_x*sin_component * gauss_component * x / 16
    deriv = 2*deriv * (sin_component * gauss_component - y)
    return np.sum(deriv)
```

### Define Gradient 
Define the gradient of the Gabor model by using the derivatives of the Gabor model with respect to each parameter (phi[0], phi[1])

```python
def compute_gradient(data_x, data_y, phi):
    dl_dphi0 = gabor_deriv_phi0(data_x, data_y, phi[0],phi[1])
    dl_dphi1 = gabor_deriv_phi1(data_x, data_y, phi[0],phi[1])
    # Return the gradient
    return np.array([[dl_dphi0],[dl_dphi1]])
```

### Minimize the Loss Function (Stochastic Gradient Descent)
Minimize the loss function by computing the gradient of a random batch of data and using the computed gradient and a constant step size to compute the stochastic descent step to adjust the parameter values and decrease the loss of the Gabor model

```python
for c_step in range (n_steps):
  # Choose random batch indices
  batch_index = np.random.permutation(data.shape[1])[0:batch_size]
  # Compute the gradient
  gradient = compute_gradient(data[0,batch_index], data[1,batch_index], phi_all[:,c_step:c_step+1] )
  # Update the parameters
  phi_all[:,c_step+1:c_step+2] = phi_all[:,c_step:c_step+1] - alpha * gradient
```

### Compute Final Loss 
Compute the loss of the Gabor model after the parameters have been adjusted with the stochastic descent step.

```python
loss =  compute_loss(data[0,:], data[1,:], model, phi_all[:,c_step+1:c_step+2])
```

### Minimize the Loss Function (Stochastic Gradient Descent + Momentum)
Minimize the loss function by computing the gradient of a random batch of data and using the computed gradient and a consistent step size to adjust the parameters. In addition, incorporate momentum into the stochastic gradient descent and use momentum to adjust the parameters as well to decrease the loss of the Gabor model.

```python
for c_step in range (n_steps):
  # Choose random batch indices
  batch_index = np.random.permutation(data.shape[1])[0:batch_size]
  # Compute the gradient
  gradient = compute_gradient(data[0,batch_index], data[1,batch_index],  phi_all[:,c_step:c_step+1])
  # Compute the momentum and incorporate it into the stochastic gradient descent step
  momentum = beta * momentum + (1 - beta) * gradient
  # Update the parameters
  phi_all[:,c_step+1:c_step+2] = phi_all[:,c_step:c_step+1] - alpha * momentum
```

Compute the loss of the Gabor model with the parameter values computed from the stochastic gradient descent with momentum

```python
loss =  compute_loss(data[0,:], data[1,:], model, phi_all[:,c_step+1:c_step+2])
```

### Minimize the Loss Function (Stochastic Gradient Descent + Nesterov Momentum) 
Minimize the loss function by computing the gradient of a random batch of data and using the computed gradient and a consistent step size to calculate the stochastic gradient descent step. In addition, incorporate the Nesterov momentum into the stochastic gradient descent step and compute the updated stochastic gradient descent step to update the parameters of the Gabor model.

```python
for c_step in range (n_steps):
  # Choose random batch indices
  batch_index = np.random.permutation(data.shape[1])[0:batch_size]
  # Compute the Nesterov momentum and incorporate it into the stochastic gradient descent step
  gradient = compute_gradient(data[0,batch_index], data[1,batch_index], phi_all[:,c_step:c_step+1] - alpha*beta*momentum)
  momentum = beta * momentum + (1-beta)*gradient
  # Update the parameters
  phi_all[:,c_step+1:c_step+2] = phi_all[:,c_step:c_step+1] - alpha * momentum
```

### Compute Final Loss
