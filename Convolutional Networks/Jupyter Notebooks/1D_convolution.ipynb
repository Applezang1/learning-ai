# 1D Convolution
This notebook aims to help gain a better understanding of 1D convolution by defining zero-padded convolution operations of different kernel, stride, and dialation values and plotting the resulting signal against the training signal

### Imports
Import the libraries needed to define the zero-padded convolution operations as well as plot the resulting signal and training signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define Signal
Define the signal, which will be the input data for the convolution operation

```python
x = [5.2, 5.3, 5.4, 5.1, 10.1, 10.3, 9.9, 10.3, 3.2, 3.4, 3.3, 3.1]
```

### Plot Signal

```python
fig,ax = plt.subplots()
ax.plot(x, 'k-')
ax.set_xlim(0,11)
ax.set_ylim(0, 12)
plt.show()
```

### Define 3, 1, 1, Convolution Operation
Define a zero-padded convolution operation with a convolution kernel size of 3, a stride of 1, and a dilation of 1

```python
def conv_3_1_1_zp(x_in, omega):
    x_out = np.zeros_like(x_in)
    x_copy = x_in.copy()
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    output_storage = []
    for i in range(1, len(x_copy)-1):
        output = omega[0]*x_copy[i-1]+ omega[1]*x_copy[i] + omega[2]*x_copy[i+1]
        output_storage.append(output)
    x_out = np.array(output_storage)
    return x_out
```

### Define Weight Values (3, 1, 1)
Define weight values for the 3, 1, 1, convolution operation

```python
omega = [0.33,0.33,0.33]
```

### Apply 3, 1, 1, Convolution Operation
Apply the convolution operation with kernel size of 3, stride of 1, and dilation of 1 onto the signal

```python
h = conv_3_1_1_zp(x, omega)
```

### Plot the Signal and Feature Map
Plot the signal (input data) and the feature map (the output of the convolution operation)

```python
fig,ax = plt.subplots()
ax.plot(x, 'k-',label='before')
ax.plot(h, 'r-',label='after')
ax.set_xlim(0,11)
ax.set_ylim(0, 12)
ax.legend()
plt.show()
```

### Define New Weight Values (3, 1, 1)
Define a new set of weight values for the 3, 1, 1, convolution operation

```python
omega = [-0.5,0,0.5]
```

### Apply 3, 1, 1, Convolution Operation
Apply the convolution operation with kernel size of 3, stride of 1, and dilation of 1 onto the signal

### Define 3, 2, 1, Convolution Operation
Define a zero-padded convolution operation with a convolution kernel size of 3, a stride of 2, and a dilation of 1

```python
def conv_3_2_1_zp(x_in, omega):
    x_out = np.zeros(int(np.ceil(len(x_in)/2)))
    output_storage = []
    x_copy = x_in.copy()
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    for i in range(1, len(x_copy)-1, 2):
        output = omega[0]*x_copy[i-1]+ omega[1]*x_copy[i] + omega[2]*x_copy[i+1]
        output_storage.append(output)
    x_out = np.array(output_storage)
    return x_out
```

### Define 5, 1, 1, Convolution Operation
Define a zero-padded convolution operation with a convolution kernel size of 5, a stride of 1, and a dilation of 1

```python
def conv_5_1_1_zp(x_in, omega):
    x_out = np.zeros_like(x_in)
    output_storage = []
    x_copy = x_in.copy()
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    for i in range(2, len(x_copy)-2, 1):
        output = omega[0]*x_copy[i-2]+ omega[1]*x_copy[i-1] + omega[2]*x_copy[i]+omega[3]*x_copy[i+1] + omega[4]*x_copy[i+2]
        output_storage.append(output)
    x_out = np.array(output_storage)
    return x_out
```

### Apply 5, 1, 1 Convolution Operation

```python
h4 = conv_5_1_1_zp(x, omega2)
```

### Plot the Signal and Feature Map

### Define 3, 1, 2 Convolution Operation
Define a zero-padded convolution operation with a convolution kernel size of 3, a stride of 1, and a dilation of 2

```python
def conv_3_1_2_zp(x_in, omega):
    x_out = np.zeros_like(x_in)
    output_storage = []
    x_copy = x_in.copy()
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    x_copy.insert(0, 0)
    x_copy.insert(len(x_copy), 0)
    for i in range(2, len(x_copy)-2):
        output = omega[0]*x_copy[i-2]+ omega[1]*x_copy[i] + omega[2]*x_copy[i+2]
        output_storage.append(output)
    x_out = np.array(output_storage)
    return x_out
```

### Define Weight Values (3, 1, 2)

```python
omega = [0.33,0.33,0.33]
```

### Apply 3, 1, 2 Convolution Operation

```python
h5 = conv_3_1_2_zp(x, omega)
```

### Plot the Signal and Feature Map

### Define 3, 1, 1, Convolution Matrix
Define a function that returns the zero-padded convolution matrix of kernel size 3, dilation 1, and step size 1

```python
def get_conv_mat_3_1_1_zp(n_out, omega):
  omega_mat = np.zeros((n_out, n_out))
  # For each output position, set the kernel weights at the correct positions
  for i in range(n_out):
    # For kernel size 3, positions are i-1, i, i+1
    for k in range(3):
      col = i + k - 1  # k=0->i-1, k=1->i, k=2->i+1
      if 0 <= col < n_out:
        omega_mat[i, col] = omega[k]
  return omega_mat
```

### Apply 3, 1, 1, Convolution Operation
Apply the 3, 1, 1, convolution operation and print the results

```python
omega = np.array([-1.0,0.5,-0.2])
h6 = conv_3_1_1_zp(x, omega)
print(h6)
```

### Apply 3, 1, 1, Convolution Matrix
Compute the 3, 1, 1 convolution matrix and compute the dot product between the convolution matrix and the input to perform the convolution operation. Print the results

```python
omega_mat = get_conv_mat_3_1_1_zp(len(x), omega)
h7 = np.matmul(omega_mat, x)
print(h7)
```
